```python
#Question-14
# Write a program to load tips dataset from seaborn and plot plot regression line between total_bill and tip 
# with 95% confidence interval.
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
df=sn.load_dataset('tips')
sns.regplot(data=df,x='total_bill',y='tip', ci=95)
plt.show()
```
