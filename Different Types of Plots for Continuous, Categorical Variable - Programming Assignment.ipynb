```python
# Question-8
# write a python program to plot two or more lines with legends different widths & colors
```

```python
import matplotlib.pyplot as pt
x=[20,40,30,50,80]
y=[70,10,20,35,80]
z=[25,75,80,30,40]
pt.plot(x,ls='dotted',lw=5,color='red',label='line 1')
pt.plot(y,ls='dashed',lw=5,color='pink',label='line 2')
pt.plot(z,ls='dashdot',lw=3,color='green',label='line 3')
pt.legend(loc='upper right')
pt.show()
```
