```python
#Question-1
# Write a program to find the number of null values in the dataset (pandas dataframe) ?
```

```python
import pandas as pd
import numpy as np
df=pd.DataFrame({'col':[43,np.nan,48,30,57,np.nan,389,np.nan]})
print(df)
print()
print(df.isnull())
print()
print(df.isnull().sum())
```

```python
#Question-2
# Write a program to create a heatmap for missing values.
```

```python
import matplotlib.pyplot as pt
import seaborn as sn
import pandas as pd
import numpy as np
df=pd.DataFrame({'col':[43,np.nan,48,30,57,np.nan,389,np.nan]})
print(df)
pt.figure(figsize=(10,8))
sn.heatmap(df.isnull().transpose(),cmap='tab20',annot=True,cbar_kws={'label':'missing Data'})
pt.show()
```

```python
#Question-14
# Write a python program to replace Nan values of area in the data set using K-NN
```

```python
import numpy as np
import pandas as pd
from sklearn.impute import KNNImputer
df=pd.DataFrame({'Col-1':[1,2,3,4],'Col-2':[5,6,7,8],'Col-3':[9,10,11,12],'Col-4':[13,14,15,16]})
print('Before Impute.....\n',df)
imputer=KNNImputer(n_neighbors=2)
imputer_data=imputer.fit_transform(df)
print()
print('After Impute.....\n',imputer_data)
```
